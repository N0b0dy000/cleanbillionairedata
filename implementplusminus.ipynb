```python
def positive_or_negative(mirror_master):
  print('start')

  available_operators = ['+', '-', '*', '/']
  last_operator = [dud for dud in mirror_master if dud in available_operators][-1]

  last_operator_location = len(mirror_master) - mirror_master[::-1].index(last_operator) - 1
  follow_operator = last_operator_location + 1

  print('safe')
  if follow_operator >= len(mirror_master):
    print('not safe', mirror_master)
    new_list = mirror_master + ['m']
    print('right')
    return new_list

  if mirror_master[follow_operator] in str([1, 2, 3, 4, 5, 6, 7, 8, 9, 0]):
    new_list = mirror_master[:follow_operator]
    new_list = new_list+['m']
    new_list = new_list+mirror_master[follow_operator:]
    # with output: display(new_list, 'after operator was a number')
    return new_list

  elif mirror_master[follow_operator] in ['m']: 
    new_list = mirror_master[:follow_operator]
    new_list = new_list+mirror_master[follow_operator+1:]

    # with output: display(new_list, 'after operator was an m')
    return new_list
```

```python
#called when looping to do calculations over a safe list
def calculation(NI, operation_string, SNI):
  NI = float(NI)
  SNI = float(SNI)
  if operation_string == "+" : calculation_output = NI + SNI
  if operation_string == "-" : calculation_output = NI - SNI
  if operation_string == "*" : calculation_output = NI * SNI
  if operation_string == "/" and SNI != 0 : calculation_output = NI / SNI
  elif operation_string == "/" and SNI == 0 : print("baka undefined")

  return calculation_output
```

```python
##should loop over interpereted_list(which has to be passed into the function) and call calculation function every 3rd value
def left_to_right_calcualtions_no_order_of_operations(list_post_interperetation):
  import numpy as np
  used_for_looping_calcutations = len(list_post_interperetation) - list_post_interperetation[::-1].index(list_post_interperetation[-1]) - 1
  track_operator_list = np.arange(1,used_for_looping_calcutations, 2)

  for track_operator in track_operator_list:
    if track_operator == 1:
      calculator_output = calculation(list_post_interperetation[track_operator - 1], list_post_interperetation[track_operator ], list_post_interperetation[track_operator + 1])
    else:
      calculator_output = calculation(calculator_output, list_post_interperetation[track_operator ], list_post_interperetation[track_operator + 1])
  return calculator_output
# calculator_result
#   with output:
#     print(calculator_output)
```
